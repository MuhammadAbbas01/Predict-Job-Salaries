```python
# L2 penalty to shrink coefficients without removing any features from the model
penalty= 'l2'
# Our classification problem is multinomial
multi_class = 'multinomial'
# Use lbfgs for L2 penalty and multinomial classes
solver = 'lbfgs'
# Max iteration = 1000
max_iter = 1000
```

```python
### BEGIN SOLUTION
from sklearn.linear_model import LogisticRegression

# Define a logistic regression model with above arguments
l2_model = LogisticRegression(random_state=42, penalty=penalty, multi_class=multi_class, solver=solver, max_iter=max_iter, class_weight='balanced')
```

```python
l2_model.fit(X_train, y_train)
```

```python
l2_preds = l2_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix,ConfusionMatrixDisplay, precision_recall_fscore_support, precision_score, recall_score
```

```python
def evaluate_metrics(yt, yp):
    results_pos = {}
    results_pos['accuracy'] = accuracy_score(yt, yp)
    precision, recall, f_beta, _ = precision_recall_fscore_support(yt, yp)
    results_pos['recall'] = recall
    results_pos['precision'] = precision
    results_pos['f1score'] = f_beta
    return results_pos
```

```python
evaluate_metrics(y_test, l2_preds)
```

The model achieved an **accuracy** of 94.52%. Class-level metrics include a **recall** range of [0.91, 1.0], a **precision** range of [0.88, 0.98], and **F1 scores** ranging from 0.94 to 0.96 across classes.

## 7.2 - KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd

# Try K from 1 to 50
max_k = 20

# Create empty lists to store evaluation metrics for each k
accuracy_scores = []
precision_scores = []
recall_scores = []
f1_scores = []

for k in range(1, max_k + 1):
    # Create a KNN classifier
    knn = KNeighborsClassifier(n_neighbors=k)

    # Train the classifier
    knn.fit(X_train, y_train.values.ravel())

    # Make predictions
    preds = knn.predict(X_test)

    # Evaluate the classifier
    accuracy = accuracy_score(y_test, preds)
    precision = precision_score(y_test, preds, average='weighted', zero_division=0)
    recall = recall_score(y_test, preds, average='weighted')
    f1 = f1_score(y_test, preds, average='weighted')

    # Store the results
    accuracy_scores.append((k, round(accuracy, 4)))
    precision_scores.append((k, round(precision, 4)))
    recall_scores.append((k, round(recall, 4)))
    f1_scores.append((k, round(f1, 4)))

```

```python
# Convert the results lists to dataframes
accuracy_results = pd.DataFrame(accuracy_scores, columns=['K', 'Accuracy'])
precision_results = pd.DataFrame(precision_scores, columns=['K', 'Precision'])
recall_results = pd.DataFrame(recall_scores, columns=['K', 'Recall'])
f1_results = pd.DataFrame(f1_scores, columns=['K', 'F1 Score'])
```

```python
f1_results
```

Using the **K-Nearest Neighbors (KNN) algorithm**, the **F1 Score** varies across different values of K, starting at **0.9488** for K=1, gradually decreasing with larger K values, and reaching **0.8848** for K=20. This trend suggests a slight performance decline as the K value increases.
